```
import pandas as pd
import altair as alt
alt.data_transformers.disable_max_rows()
campaign = pd.read_csv('https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/bank.csv')
campaignYes = campaign[campaign["y"] == "yes"]
campaignOld = campaign[campaign["age"] > 60]
campaign.head(60)
```

```
alt.Chart(campaign).mark_bar().encode(
    x = "y",
    y = 'count(y)',
    column = "job"
)
```

```
alt.Chart(campaignOld).mark_bar().encode(
    x = "y",
    y = 'count(y)',
    column = "age"
)
```

```
alt.Chart(campaign).mark_bar().encode(
    x = "y",
    y = "count(y)",
    column = "education"
)
```

```
alt.Chart(campaign).transform_joinaggregate(
    total = "count(*)"
).transform_calculate(
    pct='1 / datum.total'
).mark_bar().encode(
    alt.X('y'),
    alt.Y('count(y)', axis=alt.Axis(format='%')),
    column = "job"
)
```
